## 1. Supongamos que tomamos 10 post-its y en cada uno escribimos alguna de las letras de la palabra MATEMÁTICO. Si doblamos cada uno de los post-its y los metemos en una urna,

a. **£cuál es el espacio de probabilidad asociado a esta experiencia?**

```python
"""
El espacio de probabilidad son todas la combinaciones distintas de las letras de la palabra MATEMÁTICO.
En este caso, cuatro de las letras son indistinguibles, entonces cada palabra se repite cuatro veces.
"""


import math
import random
from itertools import permutations

palabra = "MATEMÁTICO"

# Contar las letras (con repeticiones incluidas)
letras = list(palabra)

# Espacio muestral (Ω)
# Fórmula: n! / (n1! * n2! * ... * nk!)
# donde n = total de letras, ni = repeticiones

n = len(letras)                     # total de letras = 10
nM = letras.count("M")              # cantidad de M
nT = letras.count("T")              # cantidad de T

espacio_total = math.factorial(n) // (math.factorial(nM) * math.factorial(nT))

print("Tamaño del espacio de experiencia (Ω):", espacio_total)

# Mostrar 5 combinaciones aleatorias (para ilustrar Ω)
print("\nCinco permutaciones aleatorias:")
for _ in range(5):
    muestra = "".join(random.sample(letras, len(letras)))
    print(muestra)
```

b. **£qué probabilidad hay de obtener
la palabra matemático al sacar los diez post-its de manera aleatoria?**

```python
"Para sacar la palabrá MATEMÁTICO, es solo una combinación de todo el espacio muestral Ω. Por lo que: "

prob = 1 / espacio_total

print("Probabilidad de obtener la palabra MATEMÁTICO:", f"{prob:.10f}")
```

## 3. Utilizando las probabilidades del ejercicio anterior evaluar si los siguientes eventos son dependientes o independientes
  * Obtener un numero impar
  * Que sea de color negro.

#### Definir eventos

- **Evento A:** salir un **número impar**.  
- **Evento B:** salir un **número negro**.  

#### Datos de la ruleta francesa

- Números del 1 al 36 (18 rojos, 18 negros), 0 verde.  
- Impares del 1 al 36: 1,3,5,…,35 → 18 números.   

- Probabilidades:  
$$[
P(A) = \frac{18}{36} = 0.5, \quad P(B) = \frac{18}{36} = 0.5
]$$  


¿Cuántos números son **impares y negros**?  
- En la ruleta, los negros están repartidos mitad pares y mitad impares (9 y 9).  

Entonces:  

$$
P(A \cap B) = \frac{9}{36} = 0.25
$$  



---

### Verificación de independencia
Para que A y B fueran independientes, debería cumplirse:  

$$
P(A \cap B) = P(A) \cdot P(B)
$$  

Calculamos:  

$$
P(A) \cdot P(B) = 0.5 \times 0.5 = 0.25
$$  

$$
P(A \cap B) = 0.25
$$  

**Se cumple.**  
Los eventos **“obtener un número impar”** y **“que sea de color negro”** son **independientes** en la ruleta francesa.  

---


Calcule la probabilidad condicional de obtener un número impar dado que salió el color negro P (A|B ).

Calcule la probabilidad condicional de que el número sea negro,
dado que salio un impar P (A|B ).

Graficar P (−|B ) en ambos casos.

```python
# Total de números en la ruleta (sin verde)
total = 36

# Evento A: impar (18 números)
P_A = 18 / total

# Evento B: negro (18 números)
P_B = 18 / total

# Intersección A ∩ B: impares y negros (9 números)
P_AyB = 9 / total

# Probabilidades condicionales

# P(A|B) = P(A ∩ B) / P(B)
P_A_dado_B = P_AyB / P_B

# P(B|A) = P(A ∩ B) / P(A)
P_B_dado_A = P_AyB / P_A

print("P(A) =", P_A)
print("P(B) =", P_B)
print("P(A ∩ B) =", P_AyB)

print("\nProbabilidades condicionales:")
print("P(A | B) =", P_A_dado_B)
print("P(B | A) =", P_B_dado_A)
```

En este conjunto de datos diseñado para predecir el tiempo de respuesta de un servicio de atención a emergencias, tenemos dos variables categóricas nominales, como parte de un conjunto más extenso de variables explicativas.
```python
import pandas as pd

# Lee el archivo CSV desde la URL de GitHub
url = "https://raw.githubusercontent.com/AnIsAsPe/Estadistica_y_Probabilidad_para-CD-/main/PrincipiosDeProbabilidad/DatosExamen/CategoricasExplicativasNulos.csv"
df = pd.read_csv(url)

# Muestra las primeras filas del DataFrame
display(df.head())
```

a) TipoIncidencia (Variable A). Presenta 9 categorías distintas. Algunos ejemplos del tipo de incidencia podrían ser: médica, incendio, fuga de gas, tráfico, violencia

b) TipoLugar (Variable B): Presenta 86 categorías distintas. Algunos

La Variable B (TipoLugar) contiene valores nulos. Considerando la posible relación de dependencia o independencia entre la Variable A (TipoIncidencia) y la Variable B (TipoLugar), £cuál estrategia de imputación de valores nulos consideras más adecuada para la Variable B?

Sugerimos utilizar un muestreo mediante alguna distribución que aproxime a nuestros datos o alguna correlación entre variables.

1. Análisis Previo de Dependencia
Primero, se evalúa la relación entre ambas variables mediante:

Test de Chi-cuadrado para determinar independencia estadística:

* H₀: TipoIncidencia y TipoLugar son independientes
* H₁: TipoIncidencia y TipoLugar son dependientes

**Si p < 0.05 → Dependientes**

Interpretación: Existe asociación estadísticamente significativa entre TipoIncidencia y TipoLugar.




```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# =========================================
# Análisis Previo de Dependencia
# =========================================

# Crear tabla de contingencia
contingencia = pd.crosstab(df["Tipo_Incidencia"], df["Tipo_Lugar"])

# Realizar prueba Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(contingencia)

print("Chi-cuadrado:", chi2)
print("p-valor:", p)

# Interpretación
if p < 0.05:
    print("Conclusión: Existe dependencia entre Tipo_Incidencia y Tipo_Lugar")
else:
    print("Conclusión: No hay evidencia de dependencia")

# =========================================
# Imputación Condicional Basada en Residuos de Chi-cuadrado
# =========================================
def imputar_tipo_lugar(row, contingencia, expected):
    if pd.isna(row["Tipo_Lugar"]):
        incid = row["Tipo_Incidencia"]
        # Distribución condicional ponderada por residuos (observado / esperado)
        obs = contingencia.loc[incid]
        exp = expected[contingencia.index.get_loc(incid)]
        pesos = obs / exp
        pesos = pesos / pesos.sum()  # Normalizar a probabilidades
        opciones = contingencia.columns
        return np.random.choice(opciones, p=pesos)
    else:
        return row["Tipo_Lugar"]

# Aplicar imputación
df_imputado = df.copy()
df_imputado["Tipo_Lugar"] = df_imputado.apply(
    imputar_tipo_lugar, axis=1, args=(contingencia, expected)
)

print("Imputación realizada con distribución condicional basada en residuos de Chi-cuadrado")
```

Conclusión

El análisis de dependencia mediante la prueba de Chi-cuadrado mostró que existe una asociación significativa entre Tipo_Incidencia y Tipo_Lugar (p < 0.05). Esto indica que la ocurrencia de un tipo de incidencia no es independiente del lugar donde sucede; ciertos lugares tienen mayor probabilidad de ciertos tipos de incidencias.

Para manejar los datos faltantes en Tipo_Lugar, se aplicó una imputación condicional basada en los residuos de Chi-cuadrado. Este método permite asignar valores faltantes de manera probabilística, reflejando la distribución real observada y la relación entre las variables. En otras palabras, la imputación no es aleatoria, sino que respeta la dependencia detectada entre Tipo_Incidencia y Tipo_Lugar.

** Variable Aleatoria**

- **Variable aleatoria:** $R_t$ = Retorno diario del S&P 500
- **Tipo:** Continua  
- **Definición:** Cambio porcentual diario del precio del índice:

$$
R_t = \frac{P_t - P_{t-1}}{P_{t-1}}
$$

- **Distribución aproximada:** Normal, basada en la observación empírica de los retornos históricos.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Calcular retornos diarios: r_t = (P_t - P_{t-1}) / P_{t-1}
returns = df_prices.copy()
returns = returns["Price"].pct_change()

# Parámetros para la distribución normal
mu, sigma = returns.mean(), returns.std()

# Valores para la curva teórica
x = np.linspace(returns.min()*1.2, returns.max()*1.2, 1000)
pdf = norm.pdf(x, mu, sigma)

# Gráfico
plt.figure(figsize=(8,5))
plt.hist(returns, bins=30, density=True, alpha=0.5, label="Retornos diarios (empíricos)")
plt.plot(x, pdf, 'r', lw=2, label="Distribución normal teórica")
plt.xlabel("Retorno diario")
plt.ylabel("Densidad de probabilidad")
plt.title("Espacio de probabilidad de los retornos diarios del S&P500")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Retornos diarios de tu DataFrame
returns = df_prices["Price"].pct_change().dropna()

# Parámetros de la distribución normal ajustada
mu, sigma = returns.mean(), returns.std()

# Tamaños de muestra
sample_sizes = [10, 100, 1000, 5000, 10000]

# Generar muestras aleatorias
samples = {n: np.random.normal(mu, sigma, n) for n in sample_sizes}

# Gráfico comparativo de histogramas
plt.figure(figsize=(12,8))
x = np.linspace(returns.min()*1.5, returns.max()*1.5, 1000)
plt.plot(x, norm.pdf(x, mu, sigma), 'k', lw=2, label='Distribución normal teórica')

for n, sample in samples.items():
    plt.hist(sample, bins=30, density=True, alpha=0.4, label=f'Muestra n={n}')

plt.xlabel("Retorno diario")
plt.ylabel("Densidad")
plt.title("Distribución normal teórica vs muestras aleatorias de distintos tamaños")
plt.legend()
plt.show()
```

- Para **muestras pequeñas** (por ejemplo, `n=10`), las estadísticas muestrales como la media y la desviación estándar **pueden variar significativamente** respecto a los valores poblacionales.
- A medida que **el tamaño de la muestra aumenta** (`n=1000, 5000, 10000`), las estadísticas muestrales **se aproximan cada vez más** a los parámetros poblacionales (`μ` y `σ`) de la distribución normal teórica.
- Esto ilustra directamente la **Ley de los Grandes Números**, que establece que, cuando el tamaño de la muestra crece, la media muestral converge hacia la media poblacional.

**Intervalos de confianza**
Podemos calcular un intervalo de confianza para la media de los retornos usando el Teorema Central del Límite (TCL). El TCL dice que, para muestras suficientemente grandes, la distribución de la media muestral se aproxima a una normal con media μ y desviación σ/√n.

```python
import numpy as np
from scipy.stats import norm

# Retornos diarios
returns = df_prices["Price"].pct_change().dropna()

# Tamaño de la muestra
n = len(returns)

# Media y desviación de la muestra
mean_emp = returns.mean()
std_emp = returns.std()

# Nivel de confianza (95%)
conf_level = 0.95
alpha = 1 - conf_level

# Error estándar de la media
se = std_emp / np.sqrt(n)

# Intervalo de confianza usando la normal
z = norm.ppf(1 - alpha/2)
ci_lower = mean_emp - z*se
ci_upper = mean_emp + z*se

print(f"Media empírica: {mean_emp:.6f}")
print(f"Intervalo de confianza 95%: [{ci_lower:.6f}, {ci_upper:.6f}]")
```

Nos da un rango probable para la verdadera media de los retornos diarios del S&P500.

Mientras mayor sea n, el intervalo se estrecha, reflejando la mayor precisión de la estimación gracias al Teorema Central del Límite.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Retornos diarios
returns = df_prices["Price"].pct_change().dropna()

# Tamaños de muestra
sample_sizes = [10, 100, 1000, 5000, 10000]

# Guardaremos medias y límites
means = []
ci_lowers = []
ci_uppers = []

conf_level = 0.95
alpha = 1 - conf_level

# Parámetros poblacionales
pop_mean = returns.mean()

for n in sample_sizes:
    sample = returns.sample(n, random_state=42, replace=True)
    mean_emp = sample.mean()
    std_emp = sample.std()
    se = std_emp / np.sqrt(n)
    z = norm.ppf(1 - alpha/2)
    ci_lower = mean_emp - z*se
    ci_upper = mean_emp + z*se

    means.append(mean_emp)
    ci_lowers.append(ci_lower)
    ci_uppers.append(ci_upper)

# Gráfico
plt.figure(figsize=(10,6))
plt.errorbar(sample_sizes, means, yerr=[np.array(means)-np.array(ci_lowers), np.array(ci_uppers)-np.array(means)],
             fmt='o', capsize=5, label="Media muestral ± IC 95%")
plt.axhline(pop_mean, color='r', linestyle='--', label="Media poblacional")
plt.xscale('log')
plt.xlabel("Tamaño de la muestra (log scale)")
plt.ylabel("Retorno medio")
plt.title("Convergencia de la media muestral y su IC a la media poblacional")
plt.legend()
plt.grid(True)
plt.show()
```
